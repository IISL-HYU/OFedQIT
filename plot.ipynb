```python
import pickle
import matplotlib.pyplot as plt

N = 100
L = 1
L1 = 2
L2 = 3

b = 1000
prob = 1.0
prob3 = 0.1
data = "Conduc"

with open(f"./Classification_Acc/OFedAvg_{data}_L{L}_p{prob}.pkl","rb") as f:
    ofl_sel_mse = pickle.load(f)

with open(f"./Classification_Acc/OFedQIT_{data}_L{L1}_s1_b{b}_p{prob}.pkl","rb") as f:
    ofl_sel_mse2 = pickle.load(f)
    
with open(f"./Classification_Acc/OFedAvg_{data}_L{L}_p{prob3}.pkl","rb") as f:
    ofl_sel_mse3 = pickle.load(f)

with open(f"./Classification_Acc/OFedQIT_{data}_L{L1}_s1_b{b}_p{prob3}.pkl","rb") as f:
    ofl_sel_mse4 = pickle.load(f)

    
t1 = list(range(len(ofl_sel_mse)))
t2 = list(range(len(ofl_sel_mse2)))
t3 = list(range(len(ofl_sel_mse3)))
t4 = list(range(len(ofl_sel_mse4)))
y1 = ofl_sel_mse
y2 = ofl_sel_mse2
y3 = ofl_sel_mse3
y4 = ofl_sel_mse4
# plotting the points
plt.plot(t1, y1, 'black', label=r'OFedAvg ($\mathdefault{p=1.0}$)')
plt.plot(t2, y2, 'red', label=r'OFedQIT ($\mathdefault{s=1}$, $\mathdefault{b=1000}$, $\mathdefault{L=2}$, $\mathdefault{p=1.0}$)')
plt.plot(t3, y3, 'black', label=r'OFedAvg ($\mathdefault{p=0.1}$)', linestyle='dashed')
plt.plot(t4, y4, 'red', label=r'OFedQIT ($\mathdefault{s=1}$, $\mathdefault{b=1000}$, $\mathdefault{L=2}$, $\mathdefault{p=0.1}$)', linestyle='dashed')

plt.ylim([0.5, 1]) 
plt.legend(ncol = 1)

# naming the x axis
plt.xlabel('t')
# naming the y axis
plt.ylabel('Accuracy')
plt.rcParams["font.family"] = "Times New Roman"
# giving a title to my graph
 
# function to show the plot
plt.grid()
plt.savefig(f'./Figures/Classification_{data}_OFedQIT_p={prob}_p={prob3}_b={b}_combinations.png', dpi=200, facecolor="white")
plt.show()

```
